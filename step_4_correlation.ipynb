## Removing Highly Correlated columns

```python
import db_utils as dbu
df = dbu.csv_to_df('loan_payments_filtered.csv')
```

Let's plot a correlation matrix:

```python
dfplot = dbu.Plotter(df)
```

```python
dfplot.plot_correlation_matrix()
```

Columns to remove:

- funded_amount
- funded_amount_inv
- instalment
- out_prncp_inv
- total_payment_inv
- total_rec_prncp

```python
dfplot = dbu.Plotter(df)
dfplot.plot_correlation_matrix()
```
